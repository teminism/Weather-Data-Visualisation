```python
# importing libraries used
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
```

```python
# reading dataset
weather = pd.read_csv("Test.csv")
```

```python
# checking the headings (column name) of the dataset
weather.head()
```

```python
# checking data types and value in the dataset
weather.info()
```

```python
# barplot - plotting a bar graph using those 2 columns
# focussing on the header titles in the weather dataset which are "humidity" and "temperature"
sns.barplot(weather['humidity'], weather['temperature'])
```

```python
# distplot - distinct plot of only one column for continuous data representation
# single column distribution we use distplot
sns.distplot(weather['humidity'])
```

```python
# visualising the data using more than one column
# jointplot
sns.jointplot(weather['humidity'], weather['temperature'])
```

```python
# hexagon visualisation
sns.jointplot(weather['humidity'], weather['temperature'], kind="hex")
```

```python
# kde - kernel density estimate
sns.jointplot(weather['humidity'], weather['temperature'], kind="kde")
```

```python
# pairplot is used ofr visualizing pairwise relationship in a dataset
sns.pairplot(weather[['humidity', 'temperature', 'air_pollution_index']])
```

```python
# stripplot - plotting with categorically
# shows the relationship between weather_type and temperature
sns.stripplot(weather['weather_type'], weather['temperature'])
```

```python
# jitter used to adjust the position of the plots making it easier to understand
sns.stripplot(weather['weather_type'], weather['temperature'], jitter=True)
```

```python
sns.swarmplot(weather['humidity'], weather['temperature'])
```

```python
sns.boxplot(weather['humidity'], weather['temperature'], hue=weather['weather_type'])
```

```python
sns.barplot(weather['humidity'], weather['temperature'], hue=weather['weather_type'])
```

```python
# countplot used for single columns
# counts the value of a single column
sns.countplot(weather['weather_type'])
```

```python
# point plot makes the relationship between two or more columns
sns.pointplot(weather['humidity'], weather['temperature'], hue=weather['weather_type'])
```

```python
# lmplot is for linear regression
# lmplot is most popular function to quickly plot the linear relationship between two variables
sns.lmplot(x="humidity", y="temperature", data=weather)
```

```python
sns.lmplot(x="humidity", y="temperature", hue="weather_type", data=weather)
```
